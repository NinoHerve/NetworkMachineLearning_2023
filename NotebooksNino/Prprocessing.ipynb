```python
import sys 
sys.path.append('../')

from utils.dataset import EEGDataset
from pathlib import Path


eeg_dir = Path('/home/admin/work/NetworkMachineLearning_2023/EEGDataset/')
subjects = ['sub-01']

data = EEGDataset(eeg_dir, subjects)
```

```python
import scipy.signal

fs = 250       # sampling frequency (Hz)
signal = data[100]['eeg'][50]

(f, S) = scipy.signal.periodogram(signal, fs, scaling='density')
```

```python
import matplotlib.pyplot as plt 

plt.plot(f, S)
plt.xlabel('frequency [Hz]')
plt.ylabel('PSD [V**2/Hz]')
plt.show()
```

```python
(f, S) = scipy.signal.welch(signal, fs, nperseg=200)

plt.plot(f, S)
plt.xlabel('frequency [Hz]')
plt.ylabel('PSD [V**2/Hz]')
plt.show()
```
